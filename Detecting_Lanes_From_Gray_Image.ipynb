```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image
```

# **read/display a image**

# **Convert the image to grayscale**

```python
image_gray = cv2.cvtColor(image_colored,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```
