# **RGB TO HSV**

```python
HSV_img= cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
cv2_imshow(HSV_img)
```

```python
plt.imshow(HSV_img[:,:,0])
plt.title('HUE channel')
```

```python
plt.imshow(HSV_img[:,:,1])
plt.title('Saturation channel')
```

```python
plt.imshow(HSV_img[:,:,2])
plt.title('Value channel')
```

# **Split And merge R,G,B channels**

```python
B,G,R=cv2.split(image)
B.shape
```

```python
R.shape
```

```python
zeroes = np.zeros(image.shape[:2],dtype="uint8")
cv2_imshow(cv2.merge([B,zeroes,zeroes]))
#cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_merged=cv2.merge([B,G,R])
cv2_imshow(image_merged)
```

```python
image_merged=cv2.merge([B,G+100,R])
cv2_imshow(image_merged)
```
